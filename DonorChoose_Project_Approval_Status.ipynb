```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
# Suppress warnings for clean output
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('train_data.csv')
```

```python
df.info(verbose=True)
```

```python
approval_counts = df['project_is_approved'].value_counts().sort_index()
approval_ratio = df['project_is_approved'].mean()
print(f"Approval Rate: {approval_ratio:.2%}")
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x=approval_counts.index, y=approval_counts.values, palette='Set2')
plt.title('Distribution of Project Approval Status', fontsize=14, fontweight='bold')
plt.xlabel('Approval Status (0 = Rejected, 1 = Approved)')
plt.ylabel('Number of Projects')
plt.xticks([0, 1])
for i, count in enumerate(approval_counts):
    plt.text(i, count + 50, f'{count:,}', ha='center', fontsize=10)
plt.tight_layout()
plt.show();
```

```python
labels = ['Rejected (0)', 'Approved (1)']
colors = ['salmon', 'mediumseagreen']

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(approval_counts, labels=labels, colors=colors, autopct='%1.1f%%',
        startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Project Approval Distribution', fontsize=14, fontweight='bold')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show();
```
